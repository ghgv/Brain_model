```python
class hopfield:
    def __init__(self,x1):
        (self.size1,)=x1.shape
        print(self.size1)
        self.L=np.zeros((self.size1,self.size1))
        self.W1=np.outer(x1,x1)
        self.W3=self.W1
        self.Wp=self.W1
        self.Test=self.Wp
        print("Input vector")
        print(x1)
        print("W_p matrix")
        print(self.Wp)
        print("Shape Wp:")
        print(self.Wp.shape)
        print("Norm %f " % LA.norm(self.Wp))
        
    def size(self):
        print(self.size1)
    def update(self,x2):
        W2=np.outer(x2,x2)
        (a,)=x2.shape
        if a != self.size1:
            print("Wrong input vector")
        W3=self.Wp+W2
        print("W3:")
        print(W3)
        self.L=W3.diagonal()*np.identity(self.size1)
        print("WL:")
        print(self.L)
        self.Wp=W3-self.L
        print("Wp:")
        print(self.Wp)
        print("Norm %f " % LA.norm(self.Wp))
    def forget(self,input):
        x2=input #self.convert_to_flat_inverted(input)
        (a,)=input.shape
        print(a)
        print("x2:")
        print(x2)
        W2=np.outer(x2,x2)
        if a != self.size1:
            print("Wrong input vector")
        W3=self.Wp+self.L
        #print("W3:")
        #print(W3)
        #self.L=W3.diagonal()*np.identity(self.size1)
        #print("WL:")
        #print(L)
        self.Wp=W3-W2
        print("Wp:")
        print(self.Wp)
        print("Norm %f " % LA.norm(self.Wp))
    def convert_to_flat_inverted(self,input):
        (a,)=input.shape
        if a != self.size1:
            print("Wrong input at convert to flat_inverted")
        with np.nditer(input, op_flags=['readwrite']) as it:
            for x in it:
         #       print(x, end = '')
                if x>=1:
                    x[...] = -1
         #          print("L= %i " % x)
                else:
                    x[...] = 1
          #         print("L= %i " % x)
        
        return input
    def query(self,input):
        (a,)=input.shape
        if a != self.size1:
            print("Wrong input query")
        Test=self.Wp.dot(input)
        old=copy.deepcopy(Test)
        dist=100000
        while(dist>0.1):
            with np.nditer(Test, op_flags=['readwrite']) as it:
                for x in it:
                    if x>0:
                        x[...] = 1
                    else:
                        x[...] = -1
                print(old)
                print(Test)
                dist = np.linalg.norm(old-Test)
                print("Distancia: %f " %dist)
                old=copy.deepcopy(Test)
                Test=self.Wp.dot(Test) 
        print(old)
        print("Norm %f " % LA.norm(self.Wp))
    
    def matrices(self): #Displays the NxN matrix
        W3=self.Wp
        print("NxN Matrix Wp:")
        print(W3) 
        w, v = LA.eig(W3)
        print("w:")
        print(w)
        print("v:")
        print(v)
```
